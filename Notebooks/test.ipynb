```python
import joblib, pandas as pd, numpy as np
model = joblib.load("../models/model.pkl")
df_train = pd.read_csv("../Data/Processed/customer_risk_data_final.csv")
feature_columns = df_train.drop(columns=["is_high_risk","CustomerId"]).columns.tolist()

def score(payload):
    X = pd.DataFrame([payload])
    # align columns
    for c in feature_columns:
        if c not in X.columns: X[c] = 0
    X = X[feature_columns]
    pred = model.predict(X)[0]
    proba = model.predict_proba(X)[0]
    print("classes:", model.classes_)
    print("proba vector:", proba)
    print("pred:", pred)
    print("scored_features:", X.iloc[0].to_dict())
    print()

# SAFE example (expected low-risk)
safe = {
    "Recency": 5,
    "Frequency": 40,
    "Monetary": 200000,
    "Engagement_Score": 1,
    "Channel_Diversity": 0.8,
    "Category_Diversity": 0.7
}
score(safe)

# RISKY example (expected high-risk)
risky = {
    "Recency": 300,    # very old last txn
    "Frequency": 0,    # no positive txns
    "Monetary": 0,     # no spend
    "Engagement_Score": -1,
    "Channel_Diversity": 0,
    "Category_Diversity": 0
}
score(risky)
```
